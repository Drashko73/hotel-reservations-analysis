## Hotel Reservations Data Description

```python
# Import required modules
import pandas as pd
```

```python
# Load the data
data = pd.read_csv(
    filepath_or_buffer="../data/raw/hotel_reservations.csv",
    sep=",",
    header=0
)
```

```python
# Display the first few rows of the data
data.head()
```

```python
import sys

data.info(
    verbose=True,
    buf=sys.stdout,
    memory_usage=True,
    show_counts=True
)
```

```python
# Display the number of missing values in each column
data.isnull().sum()
```

```python
# Check if there are any missing values in the data
data.isna().any()
```

```python
# Check if there are any missing values in the whole data
data.isna().sum().sum() # np.int64(0)
```

```python
# Summary statistics for the data
data.describe(
    percentiles=[0.1, 0.25, 0.5, 0.75, 0.9],
    exclude=[object]
)
```

```python
# Number of duplicate rows in the data
duplicated = data.duplicated().sum()
print(f"Number of duplicate rows in the data: {duplicated}")
```

```python
# Number of unique values in each column
unique_values = data.nunique(
    axis=0,
    dropna=False
)
print(f"Number of unique values in each column:\n{unique_values}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the boxplot for the whole data

plt.figure(figsize=(10, 5))
sns.boxplot(data=data, orient='h')
plt.xticks(rotation=90)

plt.title("Boxplot diagrams for the whole data")

plt.show()
```
